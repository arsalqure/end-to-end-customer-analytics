Run A/B test simulation and statistical tests.

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
n_A=10000
n_B=10000
p_A=0.12
p_B=0.15
count = np.array([int(n_A*p_A), int(n_B*p_B)])
stat, pval = proportions_ztest(count, np.array([n_A,n_B]), alternative='larger')
print('z-stat:', stat, 'p-value:', pval)
```
